## Physician Compare

```python
# Imports

# Transformation
import pandas as pd
import numpy as np

#Visualization
import matplotlib.pyplot as plt
import seaborn


%matplotlib inline
```

```python
file_df = pd.read_csv("./Data Sources/Physician_Compare_National_Downloadable_File.csv")
file_df.head()
```

```python
column_names = list(file_df)
column_names
```

```python
rptqm = file_df["Reported Quality Measures"].unique()
rptqm
```

```python
physbystate_df = file_df[["State", "NPI", "Reported Quality Measures"]]
physbystate_df = physbystate_df.groupby("State").count()
physbystate_df["% Reported QM"] = (physbystate_df["Reported Quality Measures"] / physbystate_df["NPI"] * 100).round()
physbystate_df.head()
```

```python
physbyschool_df = file_df[["Medical school name", "NPI"]]
physbyschool_df = physbyschool_df.groupby("Medical school name").count().sort_values(by=["NPI"], ascending=False)\
.rename(columns={"NPI" : "Count of Providers"})
physbyschool_df.head()
```
